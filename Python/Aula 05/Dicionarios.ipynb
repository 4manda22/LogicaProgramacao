```python
# Início Dicionários
    # Ordenável, mutável e não indexável.
    # Os dicionários são usados para armazenar valores de 
    # dados em pares 'chave:valor'.

umDicio =	{"marca": "Ford","modelo": "Mustang","ano": 1964}
print(umDicio)
{'marca': 'Ford', 'modelo': 'Mustang', 'ano': 1964}
# Não permite valores duplicados

umDicio =	{
  "marca": "Ford",
  "modelo": "Mustang",
  "ano": 1964,
  "marca": "Fiat",
}
print(umDicio)
{'marca': 'Fiat', 'modelo': 'Mustang', 'ano': 1964}
# Comprimento de um dicionario

umDicio =	{
  "marca": "Ford",
  "modelo": "Mustang",
  "ano": 1964,
  "marca": "Fiat",
}
print(len(umDicio))
3
# Tipos de dados aceitos

umDicio =	{
  "marca": "Ford",
  "elétrico": False,
  "ano": 1964,
  "cores": ["vermelho", "branco", "azul"]
}

print(umDicio)
{'marca': 'Ford', 'elétrico': False, 'ano': 1964, 'cores': ['vermelho', 'branco', 'azul']}
# Tipo de conjunto de dado

umDicio =	{
  "marca": "Ford",
  "modelo": "Mustang",
  "ano": 1964
}
print(type(umDicio))
<class 'dict'>
# Usando o Construtor dict()

umDicio = dict(nome = "Raquel", idade = 36, país = "Brasil")
print(umDicio)
{'nome': 'Raquel', 'idade': 36, 'país': 'Brasil'}
# Acessando um dicionário

umDicio =	{
  "marca": "Ford",
  "modelo": "Mustang",
  "ano": 1964
}
x = umDicio["marca"]
print(x)
Ford
# Pegar as Chaves

umDicio =	{
  "marca": "Ford",
  "modelo": "Mustang",
  "ano": 1964
}
x = umDicio.keys()
print(x)
dict_keys(['marca', 'modelo', 'ano'])
# Pegar os valores

umDicio =	{
  "marca": "Ford",
  "modelo": "Mustang",
  "ano": 1964
}
x = umDicio.values()
print(x)
dict_values(['Ford', 'Mustang', 1964])
# Adicionando item ao dicionario

umDicio = {
  "marca": "Ford",
  "modelo": "Mustang",
  "ano": 1964
}

x = umDicio.keys()
y = umDicio.values()

print(x)
print(y)
umDicio["cor"] = "branco"


print(x)
print(y)
print(umDicio)
dict_keys(['marca', 'modelo', 'ano'])
dict_values(['Ford', 'Mustang', 1964])
dict_keys(['marca', 'modelo', 'ano', 'cor'])
dict_values(['Ford', 'Mustang', 1964, 'branco'])
{'marca': 'Ford', 'modelo': 'Mustang', 'ano': 1964, 'cor': 'branco'}
# Ver cada item do dicionario

umDicio = {
  "marca": "Ford",
  "modelo": "Mustang",
  "ano": 1964
}
x = umDicio.items() 
print(x)
dict_items([('marca', 'Ford'), ('modelo', 'Mustang'), ('ano', 1964)])
# Verificar se a chave existe

umDicio = {
  "marca": "Ford",
  "modelo": "Mustang",
  "ano": 1964
}
if "modelo" in umDicio:
  print("Sim, 'modelo' é uma das chaves no dicionário") 
Sim, 'modelo' é uma das chaves no dicionário
# Atualizando um item

umDicio = {
  "marca": "Ford",
  "modelo": "Mustang",
  "ano": 1964
}
umDicio["ano"] = 2018
print(umDicio)
{'marca': 'Ford', 'modelo': 'Mustang', 'ano': 2018}
# Remover um item

umDicio = {
  "marca": "Ford",
  "modelo": "Mustang",
  "ano": 1964
}
umDicio.pop("modelo")
print(umDicio)
```
